```python
## There are 3 methods to create a tool in Langchain

""" | Feature                    | `@tool` Decorator | StructuredTool   | BaseTool           |
| -------------------------- | ----------------- | ---------------- | ------------------ |
| **Code complexity**        | ✅ Very low        | ⚖️ Medium        | ❌ High             |
| **Setup required**         | Minimal           | Moderate         | Full manual        |
| **Custom input schema**    | ❌ Limited         | ✅ Yes (Pydantic) | ✅ Yes (Pydantic)   |
| **Validation control**     | ❌ Basic           | ✅ Strong         | ✅ Full             |
| **Custom execution logic** | ❌ Limited         | ⚖️ Medium        | ✅ Full control     |
| **Async support**          | ❌ Not direct      | ⚖️ Partial       | ✅ Full (`_arun`)   |
| **Best for beginners**     | ✅ Yes             | ✅ Yes            | ❌ No               |
| **Agent compatibility**    | ✅ Yes             | ✅ Yes            | ✅ Yes              |
| **Production use**         | ⚖️ Rare           | ✅ Common         | ✅ Advanced systems |
| **Flexibility**            | ❌ Low             | ⚖️ Medium        | ✅ Very high        |
 """
```

```python
### How to make a custom tool in langchain

from langchain_core.tools import tool

@tool
def multiply(a: int , b: int) -> int:
    """Multiply two numbers"""
    return a*b

result = multiply.invoke({'a':3 , 'b':6})
print(result) #18
print(multiply.name) ##functon_name
print(multiply.description) ##funtion_decsription 
print(multiply.args) #args required
print(multiply.args_schema.model_json_schema())  ## json schema
```

```python
### How to make a custom tool in langchain using StructurdTool

from langchain_core.tools.structured import StructuredTool
from pydantic import BaseModel , Field

## schema defined
class MultiplyInput(BaseModel):
    a: int = Field(description="Enter first number to multiply")
    b: int = Field(description="Enter second number to multiply")

def multiply_func(a: int , b: int) -> int :
    return a*b

multiply_tool = StructuredTool.from_function(func=multiply_func, name="multiply", description='Multiply Two numbers' , args_schema=MultiplyInput)
result = multiply_tool.invoke({'a':3 , 'b':6})
print(result) #18
print(multiply_tool.name) ##function_name
print(multiply_tool.description) ##funtion_decsription 
print(multiply_tool.args) #args required
print(multiply_tool.args_schema.model_json_schema())  ## json schema
```

```python
from typing import Type
from langchain_core.tools.base import BaseTool
from pydantic import BaseModel , Field

## schema
class MultiplyInput(BaseModel):
    a: int = Field(description="Enter first number to multiply")
    b: int = Field(description="Enter second number to multiply")

#🔥 This is the key difference
#Instead of:
#wrapping a function (like StructuredTool)
#You are:
#building the tool manually
#👉 More control, more responsibility

class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"
    args_schema: Type[BaseModel] = MultiplyInput
    def _run(self , a:int , b:int ) -> int:
        return a*b
    
tool = MultiplyTool()
    
result = tool.invoke({'a':3 , 'b':6}) ## it will internally runs -run() method inside MultiplyTool
print(result) #18
print(tool.name) ##function_name
print(tool.description) ##funtion_decsription 
print(tool.args) #args Murequired
print(tool.args_schema.model_json_schema())  ## json schema
```
